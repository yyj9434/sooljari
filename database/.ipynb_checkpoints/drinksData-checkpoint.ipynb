```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://orangebottles.com/liquor"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
soup
```

```python
soup.find_all('h2')
```

```python
soup.find_all('p', 'pay no-margin')
```

```python
title_n = soup.find_all('h2')
title_n2 = soup.find_all('p', 'pay no-margin')

drinksName = [soup.find_all('h2')[n].a.string for n in range(1, len(title_n), 2)]
drinksPrice = [soup.find_all('p', 'pay no-margin')[n].string for n in range(0, len(title_n2))]
```

```python
drinksDict = {}

for i in range(0, len(title_n2)) :
    drinksDict[drinksName[i]] = drinksPrice[i]

drinksDict
```

```python
drinksDict2 = { 'drinksName' : drinksName, 'drinksPrice' : drinksPrice}
```

```python
drinksDf = pd.DataFrame(drinksDict2)
drinksDf
```

```python
drinksDf.index = drinksDf.index+1
drinksDf
```

```python
drinksDf.to_csv("drinksData.csv", mode="w", encoding='euc-kr')
```
